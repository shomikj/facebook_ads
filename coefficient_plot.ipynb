```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def formatFloat(val, fmt="%.2f"):
    ret = fmt % val
    if ret.startswith("0."):
        return ret[1:]
    if ret.startswith("-0."):
        return "-" + ret[2:]
    return ret
```

```python
groups = ["Disclosed\n&\nPersistent", "Disclosed\n&\nDisappeared", "Dark Money\n&\nPersistent", "Dark Money\n&\nDisappeared"]

coefficients = [0.310603, 0.310603+0.182274, 0.310603-0.212386, 0.310603-0.290964]
errors = [1.645*0.060146, 1.645*0.097746, 1.645*0.126540, 1.645*0.095385]
n = ['$n=388K$','$n=1.5K$','$n=20K$','$n=20K$']

fig, ax = plt.subplots(figsize=(4.5, 3))
ax.bar(groups, coefficients, color='none', yerr=errors)
ax.scatter(groups, marker='s', s=50, y=coefficients, color='black')

ax.xaxis.set_ticks_position('none')
ax.set_xticklabels(groups, fontsize=10)

for i in range(len(coefficients)):
    c = coefficients[i]
    e = errors[i]

    ax.text(i, c+e+0.02, formatFloat(c+e), horizontalalignment='center', fontsize=9, color='k')
        
    ax.text(i+0.2, c-0.03, formatFloat(c), horizontalalignment='center', fontsize=9, color='#B22222', weight="bold")
    
    ax.text(i, c-e-0.075, formatFloat(c-e), horizontalalignment='center', fontsize=9, color='k')        
    ax.text(i, -0.33, n[i], horizontalalignment='center', fontsize=9, color='#00008B')

ax.tick_params(axis='both', labelsize=9)
ax.set_ylabel('Sentiment', fontsize=9)
ax.set_xlabel('')

plt.xlim([-0.5,3.5])
plt.ylim([-0.35,0.75])
#plt.show()
fig.tight_layout()
plt.savefig('coefficients_plot.pdf', format='pdf', dpi=1000)
```
